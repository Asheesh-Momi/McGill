```python
# defining five piont stencil dervative 
def der(func,x,h):
    df=(8*func(x+h)-func(x+2*h)-8*func(x-h)+func(x-2*h))/(12*h)
    return df
def exp2(x):
    y=np.exp(0.01*x)
    return y
    
```

```python
#checking e^x for x=2 
h=np.logspace(-20, 1, num=100)
y=der(np.exp,2,h)
err=np.abs(np.exp(2)-y)
h_ideal=((45/4)*10**(-16))**(1/5)

plt.clf()
plt.xlabel("$\delta$", fontsize=14)
plt.ylabel("error", fontsize=14)
plt.title("Error vs. step size ($e^{x}$)", fontsize=14)
plt.xscale("log")
plt.yscale("log")
plt.axvline(x=h_ideal,color='r')
plt.plot(h,err)
plt.show()
print('ideal step sixe is',h_ideal,'lowest error is at', h[np.argmin(err)])
```

```python
#checking e^0.01*x for x=2 
h=np.logspace(-20, 1, num=100)
y=der(exp2,2,h)
err=np.abs(0.01*np.exp(2*0.01)-y)
h_ideal=((45/4)*10**(-16)*(1/(0.01**5)))**(1/5)

plt.clf()
plt.xlabel("$\delta$", fontsize=14)
plt.ylabel("error", fontsize=14)
plt.title("Error vs. step size ($e^{0.01x}$)", fontsize=14)
plt.xscale("log")
plt.yscale("log")
plt.axvline(x=h_ideal,color='r')
plt.plot(h,err)
plt.show()
print('ideal step sixe is',h_ideal,'lowest error is at', h[np.argmin(err)])
```

You'll note that the optimal step size in the right area

```python
#orginal unchaged from class 
def simple_integrate(fun,a,b,tol):
    x=np.linspace(a,b,5);
    dx=(b-a)/4.0;
    #np.median(np.diff(x))
    y=fun(x);
    neval=len(x) #let's keep track of function evaluations
    f1=(y[0]+4*y[2]+y[4])/6.0*(b-a);
    f2=(y[0]+4*y[1]+2*y[2]+4*y[3]+y[4])/12.0*(b-a);
    myerr=np.abs(f2-f1);
    if (myerr<tol):
        #return (f2)/1.0,myerr,neval
        return (16.0*f2-f1)/15.0,myerr,neval;
    else:
        mid=0.5*(b+a);
        f_left,err_left,neval_left=simple_integrate(fun,a,mid,tol/2.0);
        f_right,err_right,neval_right=simple_integrate(fun,mid,b,tol/2.0);
        neval=neval+neval_left+neval_right;
        f=f_left+f_right;
        err=err_left+err_right;
        return f,err,neval;
```

```python
# my recursive integartor 
def integrate(func,a,b,lim,tol):
    h=(b-a)/2
    y_old=np.zeros(3)
    y_old[0]=func(a)
    y_old[1]=func(h)
    y_old[2]=func(b)
    ans_old=(y_old[0]+4*y_old[1]+y_old[2])*(h/3) #doing out intial estimate
    count=3
    ansf,errf,count=adaptive_int(func,a,h,b,y_old,ans_old,count,tol) #calling the recursive function
    assert lim>count, "set limit is exceeded intergral may bot be evaluated to set tolerance"
    return ansf,errf,count
    
def adaptive_int(fun,a,h,b,y_old,ans_old,count,tol):
    h=h/2
    yl=np.zeros(3)
    yl[0]=y_old[0]
    yl[2]=y_old[1]
    yl[1]=fun(a+h)# copying over old values and calulating only the new one
    count= count+2
    ansl=(yl[0]+4*yl[1]+yl[2])*(h/3)# evaluating left half 
    yr=np.zeros(3)
    yr[0]=y_old[1]
    yr[2]=y_old[2]
    yr[1]=fun(b-h)# copying over old values and calulating only the new one
    count= count+2
    ansr=(yr[0]+4*yr[1]+yr[2])*(h/3)# evaluating right half 
    ans=ansl+ansr
    ans=(16.0*ans-ans_old)/15.0
    err=np.abs(ans-ans_old)
    if err>tol:
        ansl,errl,count=adaptive_int(fun,a,h,a+2*h,yl,ansl,count,tol) #recursive step if err is to high
        ansr,errr,count=adaptive_int(fun,a+2*h,h,b,yr,ansr,count,tol)
        err=errl+errr # making sure the error and ans is right 
        ans=ansl+ansr
        return ans,err,count
    else:
        return ans,err,count
```

```python
def compare(func,a,b,limit,tol):
    lz=simple_integrate(func,a,b,tol)
    my=integrate(func,a,b,limit,tol)
    print('lazy integrator has',lz[2],'calls mine has',my[2],'calls')
```

The intergral in question is 
$\frac{2\pi R^2\sigma}{4\pi\epsilon_o}\int_0^\pi \frac{(z-Rcos\theta)sin\theta}{(R^2+z^2-2Rzcos \theta)^{\frac{3}{2}}}d\theta$
for simplicty I will pick constants such that it becomes
$\int_0^\pi \frac{(z-cos\theta)sin\theta}{(1+z^2-2zcos \theta)^{\frac{3}{2}}}d\theta$

```python
z=np.linspace(0,2,101)

```

```python
def trig_stuff(x):
    y=((z-np.cos(x))*np.sin(x))/(1+z**2-2*z*np.cos(x))**(3/2)
    return y
```

My integrator

```python
Z=np.linspace(0,2,101)
E=np.zeros(101)
for i in range(101):
    z=Z[i] #going over enpough z values 
    E[i]=integrate(trig_stuff,0,np.pi,10000,10**-5)[0]
    
```

```python
plt.clf()
plt.xlabel("z", fontsize=14)
plt.ylabel("Electric field", fontsize=14)
plt.title("Addaptive integrator from Q3", fontsize=14)
plt.plot(Z,E)
plt.show()
```

You will note the the behavior is correct excluding z=1 (where the intergraor returns NaN) this is no suprise since the intergrand blows up at that piont so the intergral should not be possible to evaluate
