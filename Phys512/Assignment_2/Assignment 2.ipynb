Problem 1a

```python
x=np.linspace(0.5,1,1000) # what we are fitting 
y=np.log2(x)
```

```python
order=10 #creating values for our chi square minimization
n=len(x)
A=np.zeros([n,order+1])
xx=2*(x-x[0])/(x[-1]-x[0])-1 #rescaling the problem
A[:,0]=1.0
A[:,1]=xx
for i in range(1,order):
    A[:,i+1]=2*xx*A[:,i]-A[:,i-1] #defining the chebychev polynomials by the recursions relation
```

```python
'''This is paramter fit for the chebychev polynomial'''
#letting N be the identity (assuming noise is 1)
N=np.eye(n)
Ninv=np.linalg.inv(N)
L=np.dot(A.transpose(),np.dot(Ninv,A))
R=np.dot(A.transpose(),np.dot(Ninv,y))
m=np.dot(np.linalg.inv(L),R)# paramter values
```

```python
trunc=np.min(np.where(np.abs(m)<10**-6))# determining the value at which to truncate 
print('we truncate at the',trunc,'term')
cfit=np.dot(A[:,0:trunc],m[0:trunc])

err=np.abs(cfit-y)
print('the max error of the chebychev polynomial fit is', np.max(err))
```

We choose to truncate for values less than $10^{-6}$ since this is the error we want and the sum of all remaining values should be of order $10^{-7}$. This ended up using 8 terms 

```python
#least squares fit
m2=np.polyfit(x,y,trunc-1) 
A2=np.poly1d(m2)
pfit=A2(x)
err2=np.abs(pfit-y)
print('the max error of the least squares polynomial fit is', np.max(err2))
```

```python
#comparing the rms error
rms=np.sqrt(np.sum(np.power(err,2))/n)
rms2=np.sqrt(np.sum(np.power(err2,2))/n)
print('the root mean square error of the chebyshev fit is',rms,'while for the least squares fit its',rms2)
```

```python
"""plot of rersiduals"""
plt.plot(x,cfit-y,label='chebyshev error')
plt.plot(x,pfit-y,label='least squares error')
plt.title('comparing chebyshev and least squares residuals') 
plt.xlabel('err')
plt.ylabel('x')
plt.legend()
plt.show()
```

Like discussed in class, you can see that the Chebyshev fit has a lower maximum error (being almost half of the least-squares error) but is a worse fit overall. Its error is higher at most points, and its RMS error is 15% higher. This shows that both fits are more useful in different circumstances.

then solving the first logrthim using our checyshev fit

```python
x=np.linspace(0.01,1,1000) # what we are fitting 
y=np.log2(x)
order=100 #creating values for our chi square minimization
n=len(x)
A=np.zeros([n,order+1])
xx=2*(x-x[0])/(x[-1]-x[0])-1 #rescaling the problem
A[:,0]=1.0
A[:,1]=xx
for i in range(1,order):
    A[:,i+1]=2*xx*A[:,i]-A[:,i-1] #defining the chebychev polynomials by the recursions relation
#letting N be the identity (assuming noise is 1)
N=np.eye(n)
Ninv=np.linalg.inv(N)
L=np.dot(A.transpose(),np.dot(Ninv,A))
R=np.dot(A.transpose(),np.dot(Ninv,y))
m=np.dot(np.linalg.inv(L),R)# paramter 
trunc=np.min(np.where(np.abs(m)<10**-6))# determining the value at which to truncate 
```

```python
"""this is our function"""
def log2(X):
    y1,y2=np.frexp(X) #decomposing our imput
    a=np.zeros(trunc)
    a[0]=1.0
    yy=2*(y1-x[0])/(x[-1]-x[0])-1 # rescaling the imput
    a[1]=yy
    for i in range(1,trunc-1):
        a[i+1]=2*yy*a[i]-a[i-1] 
    y=np.dot(a,m[0:trunc])+y2
    return y
```

We would fit this function with something of the form
$ae^{-b(x-c)}+d$
my intial guesses would be
a=0.2, b=50, c=the first data piont of x, d=1

this function has a nonlinear some of its pramaters (b,c) so it is nonlinear

```python
#intial guess
a=0.2
b=50
c=xt[0]
d=1
yi=a*np.exp(-b*(xt-c))+d # intial fit paramters
```

```python
plt.plot(xt,yi,label='fit')
plt.plot(xt,yt,label='data')
plt.legend()
plt.show()
```

2b

$\frac{\partial y}{\partial a}=e^{-b(x-c)} \\ \frac{\partial y}{\partial b}=-a(x-c)e^{-b(x-c)} \\ \frac{\partial y}{\partial d}=1$

To estimate the noise we will be looking at the background fluctations and assume that noise is uncorrelated 

```python
plt.plot(x[0:3000],y[0:3000])
plt.show()
```

To estimate $\sigma$ we are looking at the noise on an area the is before the flare wem then assume the spread of the data is $6\sigma$

```python
"""setting prelimanry values for our nonlinear chi square"""
yf=yi # intial fit paramters
sigma=0.02/6 #standard deviation
n=len(yf)
m=[a,b,d] #paramater array
N=np.eye(n)*sigma**2
Ninv=np.eye(n)/sigma**2
dA=np.zeros([n,3]) # derrivtive array
dA[:,0]=np.exp(-m[1]*(xt-c))
dA[:,1]=-y[0]*(xt-c)*np.exp(-y[1]*(xt-c))
dA[:,2]=1
max_steps=30
chi_sq=np.zeros(max_steps+1)
chi_sq[0]=np.dot((yt-yf).transpose(),np.dot(Ninv,yt-yf))
```

```python
"""implementing newtons method stoping if the change in chi square is less than 1"""
for i in range(0,max_steps-1):
    yf=m[0]*np.exp(-m[1]*(xt-c))+m[2]
    GX=np.dot(dA.transpose(),np.dot(Ninv,dA))
    K=np.dot(dA.transpose(),np.dot(Ninv,yt-yf))
    m=m+np.dot(np.linalg.inv(GX),K)
    dA[:,0]=np.exp(-m[1]*(xt-c))
    dA[:,1]=-y[0]*(xt-c)*np.exp(-y[1]*(xt-c))
    dA[:,2]=1
    chi_sq[i]=np.dot((yt-yf).transpose(),np.dot(Ninv,yt-yf))
    if np.abs(chi_sq[i-1]-chi_sq[i])<1:
        print('stopped after ',i,'iterations')
        break
if chi_sq[-1]!=0:
    print('reached iteration limit of', maxsteps)
    
print('the best fit paramters are a=',m[0],'b=',m[1],'d=',m[2],'& chi squared is',chi_sq[i])
m_min=np.copy(m)


```
